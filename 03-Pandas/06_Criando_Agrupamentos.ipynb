## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo', 'sum': 'Soma'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

## Estatísticas Descritivas Pt. 2

```python
dados_imoveis = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados_imoveis
```

```python
dados_imoveis.shape[0]
```

```python
dados_imoveis.info()
```

```python
dados_imoveis['Valor m2'] = (dados_imoveis['Valor'] / dados_imoveis['Area']).round(2)
```

```python
dados_imoveis
```

```python
dados_imoveis['Valor m2'].info()
```

```python
dados_imoveis.dropna(inplace=True)
```

```python
dados_imoveis.info()
```

```python
grupo_tipo_imoveis = dados_imoveis.groupby('Tipo')
```

```python
grupo_tipo_imoveis
```

```python
grupo_tipo_imoveis['Valor m2'].mean().round(2)
```

```python
grupo_tipo_imoveis['Valor m2'].describe().round(2)
```

```python
colunas = ['mean', 'count', 'min', 'max', 'std', 'sum']
```

```python
colunas_renomeadas = {'mean': 'Média', 'count': 'Quantidade', 'min': 'Mínimo', 'max': 'Máximo', 'std': 'Desvio Padrão', 'sum': 'Soma'}
```

```python
grupo_tipo_imoveis['Valor m2'].agg(colunas).rename(columns=colunas_renomeadas)
```
